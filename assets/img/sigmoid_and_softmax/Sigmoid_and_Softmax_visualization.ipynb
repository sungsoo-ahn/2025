```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
x = np.random.uniform(-1000, 1000, 1)
x
#w=np.meshgrid(np.linspace(0, 1, 10), np.linspace(0, 1, 10))
```

```python
data_x, data_y, data_z =[], [], []
for r in np.arange(0,1, 0.001):
    for alpha in np.arange(-1000,1000, 10):
            data_x += [r]
            data_y += [alpha]
            data_z += [r*x - alpha*(r*np.log(r+1e-8) + (1-r)*np.log(1-r+1e-8)) ]
```

```python
fig = plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection='3d')
ax.scatter(data_x, data_y, data_z, c=data_z)
ax.view_init(azim=-45,elev=30)
plt.title(r'$wx - w\log\left(w\right) - \left(1-w\right)\log\left(1-w\right)$'+ '\n'+ r'$x = ' +str(x[0].round(2))+'$')
ax.set_xlabel(r'$w$')
ax.set_ylabel(r'$\alpha$')
#ax.set_zlabel(r'$wx - w\log\left(w\right) - \left(1-w\right)\log\left(1-w\right)$')
plt.savefig('sigmoid_neg_entropy.png', dpi=6e2, bbox_inches='tight', transparent = True)
plt.show()
```

```python
x = np.random.uniform(-1000, 1000, 3)
x
#w=np.meshgrid(np.linspace(0, 1, 10), np.linspace(0, 1, 10))
```

```python

data_x, data_y, data_z, data_w =[], [], [], []
for r in np.arange(0,1, 0.001):
    for c in np.arange(0,1, 0.001):
        for alpha in np.arange(-1000,1000, 10):
                if r+c > 1:
                    continue
                data_x += [r]
                data_w += [c]
                data_y += [alpha]
                data_z += [r*x[0] + c*x[1]+ (1-r-c)*x[2] + alpha*(r*np.log(r+1e-8) + c*np.log(c+1e-8) + (1-r-c)*np.log(1-r-c+1e-8)) ]
```

```python

fig = plt.figure(figsize=(10,10))
ax=fig.add_subplot(projection='3d')
ax.scatter(data_x, data_w, data_z, c=data_z)
ax.view_init(azim=-45,elev=30)
plt.tight_layout(pad=2)
plt.title(r'$\langle \mathbf{w}, \mathbf{x}\rangle + \alpha\sum_{i=1}^d \mathbf{w}_i\log\left(\mathbf{w}_i\right)$'+ '\n'+ r'$\mathbf{x} = ['+ '{:.2f}'.format(x[0])+', '+'{:.2f}'.format(x[1])+', '+'{:.2f}'.format(x[2])+']$')
ax.set_xlabel(r'$\mathbf{w}_1$')
ax.set_ylabel(r'$\alpha$')
ax.set_zlabel(r'$\mathbf{\hat{p}}_i$')
plt.savefig('softmax_neg_entropy.png', dpi=6e2, bbox_inches='tight', transparent = True)
plt.show()
```
